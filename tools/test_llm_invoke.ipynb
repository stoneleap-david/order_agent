```python



# 1.导入相关依赖
from langchain_openai import ChatOpenAI

from  dotenv import load_dotenv
load_dotenv()

import  os

# 2.获取api_key和base_url
api_key= os.getenv("DASHSCOPE_API_KEY")
base_url=os.getenv("DASHSCOPE_API_BASE")

# 3.实例化模型
llm = ChatOpenAI(model_name="qwen2.5-72b-instruct",openai_api_key=api_key ,openai_api_base=base_url)

# 4.调用模型
llm.invoke("请给我讲一个幽默的笑话")

```
